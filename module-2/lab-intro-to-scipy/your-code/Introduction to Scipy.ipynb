# The stats Submodule

The stats submodule contains a large portion of statistical functions as well as probability distributions.
Distributions

The first distribution we will look at is the normal distribution. We will pass a collection of numbers to this function and we will get the probability density function. You may be familiar with the normal distribution has having a bell shape. For each value that we will pass to the function, we will get the probability of that value occurring. The closer we are to the mean, the more likely the value. This is what gives us the bell shape.

We can look at the cumulative distribution function and find the probability of x > 0 for a normal distribution with a mean of zero and a standard deviation of one.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

normal = norm()
norm.cdf(0)
```

In the example below, we will draw the bell curve. The mean is zero and the standard deviation is 1 (these are the defaults for this function). Therefore, the closer we are to zero, the more likely the value.

```python
# The code below generates numbers between -1 and 1 in 0.1 increments
x = np.arange(-1, 1.1, 0.1)
# Now we will initialize the normal distribution object
normal = norm()
# Below we will use matplotlib to plot the bell curve. We will generate the probability density fuction
# (pdf) for the values we generated in x
plt.plot(x, norm.pdf(x))
```

The stats submodule contains many other distributions like the binomial distribution, the poisson distribution, and the exponential distribution. All of these are common distributions in statistics.



# The linalg Submodule

This submodule helps us perform linear algebra calculations on our data. Typically, we pass numpy arrays to functions in this submodule. Using linear algebra, we can transpose and invert matrices. We can also solve a set of linear equations and find the eigenvalues of a matrix.

For example, in the code below, we will solve a set of linear equations.

We have the two equations: 5x + 6y = 10 and 2x + 4y = 8. These two equations can be separated to one matrix for the coefficients on the left side and one for the constants on the right. We will pass these two matrices to the solve function to find the possible values for x and y.

```python
from scipy.linalg import solve

a = np.array([[5, 6], [2, 4]])
b = np.array([10, 8])
x = solve(a, b)
x
```

This means that the solutions are x = -1, y = 2.5.

# The interpolate Submodule

This submodule contains functions for interpolating between two points using different algorithms. Interpolating between two points can be particularly useful when we have two points that are too far apart, and we would like to estimate what is potentially happening between the points. One such function is the interp1d function. This function uses multiple algorithms to interpolate between two points. The two most common options for interpolation are linear and spline interpolation. Let's look at an example of linear interpolation.

```python
from scipy.interpolate import interp1d

# We are creating a dataset where the x values are numbers 0 through 10 
# and the y values are the cosine of the x values
x = np.arange(0, 11, 1)
y = np.cos(x)

# We generate an interpolation function and then apply that function to the new data and plot
f = interp1d(x, y)
xnew = np.arange(0, 10.25, 0.25)
plt.plot(xnew, f(xnew))
```

We can also interpolate using a quadratic spline.

```python
f = interp1d(x, y, 'quadratic')
xnew = np.arange(0, 10.25, 0.25)
plt.plot(xnew, f(xnew))
```

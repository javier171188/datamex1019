```python
def get_coor(s):
    
    lst = s.split(' ')
    lst[1]=lst[1].replace('(', '')
    lst[2]=lst[2].replace(')', '')
    #print([lst[1],lst[2]])
    return lst[1],lst[2]
```

```python
coor_esc =[]
for i in range(len(d['data'])):
    #print(d['data'][i][-1])
    coor_esc.append(get_coor(d['data'][i][-1]))
```
